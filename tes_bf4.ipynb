```python
import pandas as pd

df = pd.read_csv("Data/PV web_negative.csv")
df['URL'] = df['URL'].apply(lambda url: 'https://' + url)
urls = df['URL'].to_list()
```

```python
import cfscrape 
from bs4 import BeautifulSoup

teks = []
for url in urls :
    try :
        scraper = cfscrape.create_scraper() 
        scraped_data = scraper.get(url) 

        soup = BeautifulSoup(scraped_data.text, 'html.parser')
        text = soup.get_text()
        text = " ".join(text.split())
        teks.append(text)
        print("Success: ", url)
    except Exception as e :
        print("Error : ",url)
        continue
```
